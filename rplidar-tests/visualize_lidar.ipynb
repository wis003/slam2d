```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def visualize_lidar(path, instance):
  data = pd.read_csv(path, delimiter = ',')

  array = np.fromstring(data['ranges'][instance][1:-1], sep=',')
  array[np.where(array == np.inf)] = 20.0
  theta = np.arange(0, 2*np.pi, 2*np.pi/720)
  for i, ele in enumerate(theta):
    if ele > 4.88692 and ele < 5.06145:
      print(array[i])

  fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
  ax.plot(theta, array)
  ax.set_rmax(2.0)
  # ax.set_rticks([0.25, 0.5, 0.75, 1])  # Less radial ticks
  plt.show()

visualize_lidar('topfovfar/scan.csv', 0)
visualize_lidar('topfovfar/scan.csv', 10)
visualize_lidar('topfovfar/scan.csv', 20)
```
